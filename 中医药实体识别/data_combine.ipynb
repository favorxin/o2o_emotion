```python
import pandas as pd
import os
import json
```

```python
train_file_dir = './round1_train/train/'
train_file_names = os.listdir(train_file_dir)
train_file_names
```

```python
train_file_names[4][3:]
```

```python
train_content_list = []
train_label_list = []
for idx, file_name in enumerate(train_file_names):
    if file_name.split('.')[1] == 'txt':
        f = open(train_file_dir + file_name, encoding='utf-8')
        cur_txt = f.read()
        cur_txt_1 = cur_txt.split(' ')
        txt_all = ''
        for data in cur_txt_1:
            if data != ' ':
                txt_all += data
        train_content_list.append(txt_all)
        f.close()
    elif file_name.split('.')[1] == 'ann':
        f = open(train_file_dir + file_name, encoding='utf-8')
        cur_label = f.read()
        cur_label_1 = cur_label.split('\n')
        train_label_list.append(cur_label_1)
        f.close()
    else:
        print(file_name)
#     if len(train_content_list) == 5 & len(train_label_list) == 5:
#         break
all_train_content_list = train_content_list
all_label_content_list = train_label_list
```

```python
label_class_dict = {}
for label_content_list in train_label_list:
    for cur_infor in label_content_list:
        cur_list = cur_infor.split('\t')
        cur_class = cur_list[1].split(' ')
        if cur_class[0] not in label_class_dict:
            label_list = []
            label_list.append(cur_list[2])
            label_class_dict[cur_class[0]] = label_list
        else:
            if len(label_class_dict[cur_class[0]]) < 5:
                label_class_dict[cur_class[0]].append(cur_list[2])
```
